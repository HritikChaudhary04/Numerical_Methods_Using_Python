## Implement Euler Method For differential equation by Hritik Chaudhary CM-2021

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def fun(x,y):
    z=3-2*x-y/2
    return z

def exact(x):
    return(14-4*x-13*np.exp(-x**2))
def euler(x,y,h):
    x_values=np.arange(x[0],x[1]+h,h)
    y_values=[y]
    
    for i in range(1,len(x_values)):
        y_0=y_values[i-1]+h*fun(x_values[i-1],y_values[i-1])
        y_values.append(y_0)
     
    return x_values,y_values

x=[0,2]
y=1
h=[0.5,0.1,0.01]

x_exact=np.linspace(x[0],x[1],100)
y_exact=exact(x_exact)
plt.plot(x_exact,y_exact,label='Exact Solution',color='r',linestyle='--')
for step in h:
    x_euler,y_euler=euler(x,y,step)
    plt.plot(x_euler,y_euler,label=f'for h = {step} the euler graph')


plt.title('Euler\'s Method for ODE')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()


        
    
```
